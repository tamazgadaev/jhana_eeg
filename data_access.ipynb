# How to load BIDS format data

```python
import mne
from mne_bids import read_raw_bids, find_matching_paths
```

```python
root_path = '/path/to/data/'
```

```python
bpaths = find_matching_paths(root_path, extensions=".edf", suffixes="eeg")
path = bpaths[0]
raw = read_raw_bids(bids_path=path, verbose=False)
annot_df = raw.annotations.to_data_frame()
```
